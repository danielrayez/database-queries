```python
url = 'https://www.resultados-futbol.com/Metropolitanos'
r = requests.get(url)
ccs_soup = BeautifulSoup(r.text, 'lxml')
```

```python
equipo = str(url[34:])
# equipo
```

```python
lista_cruda_plantilla = ccs_soup.find_all('tr', attrs= {'itemprop':'employee'})

type(lista_cruda_plantilla)
```

```python
#EXTRAER NOMBRE
nombre = lista_cruda_plantilla[19].find('span').text

nombre

```

```python
#EXTRAER NUMERO EN EL EQUIPO

numero_jugador = lista_cruda_plantilla[0].find('td', attrs= {"class":"num"}).text
numero_jugador
```

```python
# EXTRACCIÓN DE LA EDAD

edad_jugador = lista_cruda_plantilla[0].find('td', attrs={'itemprop':"birthDate"}).text
edad_jugador
```

```python
#EXTRACCION DE LA PROCEDENCIA (PAÍS)

tag_nacionalidad = lista_cruda_plantilla[19].find('td', attrs= {'itemprop':'nationality'})
str_nacionalidad = str(tag_nacionalidad)

nacionalidad = str_nacionalidad[104:106]
nacionalidad
```

```python
# EXTRACCIÓN DEL PESO Y LA ALTURA. 1ER PASO

altura_peso = lista_cruda_plantilla[0].find_all('td', attrs= {'class':'dat'}, limit=2)
type(altura_peso)
```

```python
# EXTRACCION DE LA ALTURA
altura= altura_peso[0].text
altura
```

```python
# EXTRACCION DEL PESO
peso= altura_peso[1].text
peso
```

```python
# EXTRACCION DE POSICIÓN DE UN ATRIBUTO COLSPAN

posicion = lista_cruda_plantilla[0].find_previous_sibling('tr', attrs = {'class':'first-child'}).get_text(strip = True)
posicion
```

```python
#PREVISUALIZACIÓN DE LA LISTA DE DATOS DEL JUGADOR
datos_jugador = [nombre, edad_jugador, nacionalidad, altura, peso]
datos_jugador
```

```python
#BUCLE DE EXTRACCION DE DATOS CON CICLO FOR
plantilla = []

contador = 0
for item in lista_cruda_plantilla:
    
    #equipo
    equipo
    #posicion
    posicion = lista_cruda_plantilla[contador].find_previous_sibling('tr', attrs = {'class':'first-child'}).get_text(strip = True)
    #nombre
    nombre = lista_cruda_plantilla[contador].find('span').text
    #numero
    numero_jugador = lista_cruda_plantilla[contador].find('td', attrs= {"class":"num"}).text
    #edad
    edad_jugador = lista_cruda_plantilla[contador].find('td', attrs={'itemprop':"birthDate"}).text
    #nacionalidad
    tag_nacionalidad = lista_cruda_plantilla[contador].find('td', attrs= {'itemprop':'nationality'})
    str_nacionalidad = str(tag_nacionalidad)
    nacionalidad = str_nacionalidad[104:106]
    #Altura y peso
    altura_peso = lista_cruda_plantilla[contador].find_all('td', attrs= {'class':'dat'}, limit=2)
    altura= altura_peso[0].text
    peso= altura_peso[1].text

    #agregar nuevo elemento a la lista
    plantilla.append(equipo)
    plantilla.append(numero_jugador)
    plantilla.append(posicion)
    plantilla.append(nombre)
    plantilla.append(edad_jugador)
    plantilla.append(nacionalidad)
    plantilla.append(altura)
    plantilla.append(peso)
    
    contador = contador +1
    
    
```

```python
#VISTA DE LA PLANTILLA COMO DATOS CRUDOS
plantilla
```

```python
# SEPARANDO LA LISTA EN SUBLISTAS POR JUGADOR
n=8

plantilla_final =[plantilla[i:i + n] for i in range(0, len(plantilla), 
                                                    n)]
plantilla_final
```

```python
#PRESENTACIÓN DE PLANTILLA COMO DATAFRAME
import pandas as pd

plantilla_df = pd.DataFrame(plantilla_final)
plantilla_df
```

```python
#RENOMBRE DE LAS COLUMNAS
plantilla_df.columns= ['equipo','numero','posicion','nombre','edad','nacionalidad','altura','peso']
plantilla_df
```

```python
#Chequeando el tipo de datos
plantilla_df.info()
```

#### cambiando el tipo de datos a las columnas numericas

```python
plantilla_df['altura'] = pd.to_numeric(plantilla_df['altura'], errors= 'coerce')
```

```python
plantilla_df['edad'] = pd.to_numeric(plantilla_df['edad'], errors= 'coerce')
```

```python
plantilla_df['peso'] = pd.to_numeric(plantilla_df['peso'], errors= 'coerce')
```
